```python
import pandas as pd

# Create a sample DataFrame
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age': [23, 30, 19, 45]
}
df = pd.DataFrame(data)

# Filter rows where age > 25
ages = df[df['age'] > 25]
print(ages)
```

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'a': [1, 2, 3, 4, 5],
    'b': [10, 20, 30, 40, 50]
}
df = pd.DataFrame(data)

# Add a new column 'total' as the sum of 'a' and 'b'
df['total'] = df['a'] + df['b']

# Display the first few rows
print(df.head())
```

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'category': ['A', 'B', 'A', 'C', 'B', 'C'],
    'sales': [100, 200, 150, 300, 250, 400]
}
df = pd.DataFrame(data)

# Group by 'category' and sum the 'sales'
res = df.groupby('category')['sales'].sum()

# Print the result
print(res)
```
